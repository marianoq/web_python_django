```python
import pandas as pd
```

```python
data = pd.read_csv('./datasets/ZoneA.dat')
```

```python
#extraer coordenadas
coordenadas = []

for file in range(9,len(data)):
    x = (data.iloc[file,0]).split(' ')[0]
    y = (data.iloc[file,0]).split(' ')[1]
    z = (data.iloc[file,0]).split(' ')[4]
    coordenadas.append([x,y,z])
```
